```python
from statsmodels.tsa.stattools import adfuller
test_result=adfuller(data_set['Close'])
print("ADF p-value:", test_result[1])

if test_result[1]<=0.5:
    print("The data is stationary")
else:
    print("The data is non stationary") 
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(data_set['Close'],lags=30)
plt.show()

plot_pacf(data_set['Close'],lags=30)
plt.show()
```

```python
data_set['diff']=data_set['Close'].diff().diff()
data_set['diff'].dropna(inplace=True)
data_set['diff'].plot(title="First Difference")
```

```python
data_set['diff1'] = data_set['Close'].diff()
data_set['diff1'].plot(title="second Difference")
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(data_set['diff'].dropna())
print("ADF p-value:", result[1])

if result[1]<=0.5:
    print("the data is stationary")
else:
    print("The data is non stationary")    
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data_set['diff'].dropna(), lags=30)
plot_pacf(data_set['diff'].dropna(), lags=30)
```

```python
import statsmodels.api as sm
import warnings

p , d, q = 2, 2, 1

model = sm.tsa.statespace.SARIMAX(data_set['Close'],order=(p, d, q),
                                  seasonal_order=(p, d, q, 12))
model = model.fit()
print(model.summary())
forecast = model.forecast(steps=30)
```

```python
plt.figure(figsize=(12,4))
forecast.plot(c='green')
```

```python
last_date = data_set.index[-1]
future_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30)
forecast_series = pd.Series(forecast.values, index=future_index)
plt.figure(figsize=(12,4))
plt.plot(data_set['Close'], label='Actual')
plt.plot(forecast_series, label='Forecast', color='orange')
plt.title("Actual vs Forecast")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

start = data_set.index[100]       
end = data_set.index[-1]
predicted = model.predict(start=start, end=end, dynamic=False)

actual = data_set.loc[start:end, 'Close']

mse = mean_squared_error(actual, predicted)
print("MSE:", mse)

plt.figure(figsize=(14,6))
plt.plot(actual, label='Actual', color='red')
plt.plot(predicted, label='Predicted', color='blue')
plt.title('Actual vs Predicted Closing Prices (In-sample)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
