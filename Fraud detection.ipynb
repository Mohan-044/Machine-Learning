```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
encoder=LabelEncoder()
data_set['en_nameOrig']=encoder.fit_transform(data_set['nameOrig'])
data_set['en_nameDest']=encoder.fit_transform(data_set['nameDest'])

OneHotEncoder_obj=OneHotEncoder()
data_set1=OneHotEncoder_obj.fit_transform(data_set[['type']]).toarray()
data_set2 = pd.DataFrame(data_set1)
```

```python
data_set3=pd.concat([data_set,data_set2],axis=1)
data_set3.drop(columns="type",inplace=True)
data_set3.drop(columns="nameOrig",inplace=True)
data_set3.drop(columns="nameDest",inplace=True)
data_set3.drop(columns=0,inplace=True)
```

```python
data_set3.drop(columns='en_nameOrig',inplace=True)
data_set3.drop(columns='isFlaggedFraud',inplace=True)
data_set3.drop(columns='en_nameDest',inplace=True)
```

```python
x=data_set3.drop(columns='isFraud')
y=data_set3['isFraud']
```

```python
data_set3.dtypes
```

```python
x.isna().sum()
```

```python
x.dropna(inplace=True)
y.dropna(inplace=True)
```

```python
print(type(y))
```

```python
from imblearn.over_sampling import SMOTE
x.columns = x.columns.astype(str)
smote=SMOTE(sampling_strategy="minority")
xs,ys=smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xs,ys,test_size=0.2,random_state=0,stratify=ys)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(
    n_estimators=200,         # number of trees
    max_depth=6,              # depth of each tree
    learning_rate=0.1,        # step size
    subsample=0.8,            # fraction of samples
    colsample_bytree=0.8,     # fraction of features
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
